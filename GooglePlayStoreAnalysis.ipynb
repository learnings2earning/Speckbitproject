```python
installnumlist = list(data.Installs)#Extracting the install column from data
for i in installnumlist:# Replacing '0+' and '0' with '1000+' in install column
    if i == '0' or i== '0+' or i== 'Free':
        iniind = installnumlist.index(i)
        installnumlist[iniind]='1000+'
installnumlist1=list()#List to store install numbers not having '+' sign
for i in installnumlist:
   length=len(i)
   a= i[:(length-1)]
   installnumlist1.append(a)
installnumlist2 = list()#Creating and storing the install without comma
for j in installnumlist1 :
    a = (j.replace(',', ''))
    installnumlist2.append(a)
installnumlist3 = list()#Creating and storing the install column in integer datatype
try:
    for k in installnumlist2:
        b = (int(k))
        installnumlist3.append(b)#All installation numbers in integer data type are in this list(installnumlist3)
except:
    print('Error')


```

```python
all_installs=dict()#Storing all installation numbers for each category in dictionary all_installs
allmax_installs=dict()#Storing maximum installation number in each category
data1 = pd.concat([data.Category,pd.DataFrame(installnumlist3, columns=['Installs'])],axis=1)
data1.head()

for i in range(len(data1)):
    if (data1.iloc[i,0]) != '1.9':
        all_installs[data1.iloc[i,0]] = []

for i in range(len(data1)):#Finding the maximum installation number in each category
    if (data1.iloc[i,0]) != '1.9': 
        all_installs[data1.iloc[i,0]].append(data1.iloc[i,1])
max=0
for j in all_installs.keys():
    l = all_installs[j]
    for k in l:
        if k > max:
            max=k
    allmax_installs[j]=max
    max=0
print(allmax_installs)

```
